```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/BAC.csv')  # Modify this based on the dataset structure

# Display first few rows
display(df.head())

# Example 1: Bar Chart
plt.figure(figsize=(8,5))
# Replace 'Column_Name' with an actual column name from your DataFrame, e.g., 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'
df['Open'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# Example 2: Line Graph
plt.figure(figsize=(8,5))
# Replace 'Date' and 'Value' with actual columns
# Assuming 'Date' is a column in your DataFrame and you want to plot 'Close' values over time
plt.plot(df['Date'], df['Close'], marker='o', linestyle='-')
plt.title('Line Graph Example')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()

# Example 3: Scatter Plot
plt.figure(figsize=(8,5))
# Replace with actual columns
# For example, to plot 'Open' against 'Close'
sns.scatterplot(data=df, x='Open', y='Close')
plt.title('Scatter Plot Example')
plt.show()

# Example 4: Histogram
plt.figure(figsize=(8,5))
# Replace 'Numeric_Column' with an actual numeric column, e.g., 'Volume'
df['Volume'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.show()
```
